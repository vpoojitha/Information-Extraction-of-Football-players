# Task 1 
Creating  a function that takes each document and performs:
1) sentence segmentation 2) tokenization 3) part-of-speech tagging



```python
def ie_preprocess(document):
        
    # Sentence tokenize the document
    pos_sentences = nltk.sent_tokenize(document)
    
    # Word Tokenize each sentence
    pos_sentences = [nltk.word_tokenize(sent) for sent in pos_sentences]
    
    # POS tagging for each tokens
    pos_sentences = [nltk.pos_tag(sent) for sent in pos_sentences]
    
    return pos_sentences
```

# Task 2
Creating a function that will take the list of tokens with POS tags for each sentence and returns the named entities (NE). 



```python
from nltk import ne_chunk
            
def named_entity_finding(pos_sent):
    
    
    tree = nltk.ne_chunk(pos_sent, binary=True)
    #print(tree)
    #traversing the tree object to get to the NEs
    '''(NE Cristiano/NNP Ronaldo/NNP)
    dos/NN
    (NE Santos/NNP Aveiro/NNP)
    ,/,
    (NE ComM/NNP)
    ,/,
    (NE GOIH/NNP)
    (/(
    born/VBN
    5/CD
    February/NNP
    1985/CD
    )/)
    is/VBZ
    a/DT
    (NE Portuguese/JJ)
    professional/JJ
    footballer/NN
    who/WP
    plays/VBZ
    for/IN
    (NE Spanish/JJ)
    club/NN
    (NE Real/NNP Madrid/NNP)
    and/CC
    the/DT
    (NE Portugal/NNP)
    national/JJ
    team/NN
    ./.)'''


    named_entities = []

    # choosing the subtrees with NE as label
    
    for subtree in tree.subtrees(): 
        #print(subtree)
        if subtree.label() == 'NE': 
            entity = "" 
            for leaf in subtree.leaves(): 
                entity = entity + leaf[0] + " "

            # Append the resultant entity into Output list
            named_entities.append(entity.strip())
    return named_entities

post_sents=ie_preprocess(list_of_doc[0]) 

# Check the output

print("Output ", named_entity_finding(post_sents[0]))
```

```python
def NE_flat_list_fn(pos_sents): 
    NE=[]
    for pos_sent in pos_sents:
        NE.append(named_entity_finding(pos_sent))
    #Single line code here. Flatten the list of lists to the single list NE_flat_list
    NE_flat_list = [item for sublist in NE for item in sublist]
    return(NE_flat_list)

post_sents = []

for doc in list_of_doc:
    post_sents.append(ie_preprocess(doc))
 
NE_flat_list_fn([item for sublist in post_sents for item in sublist])
```

# Task 4 

Creating a functions to extract the name of the player, country of origin and date of birth as well as the following relations: team(s) of the player and position(s) of the player.


Reference: https://docs.python.org/3/howto/regex.html

```python
from nltk.sem import extract_rels,rtuple, relextract
from nltk.corpus import stopwords

def name_of_the_player(doc):
       
    post_sent=ie_preprocess(doc) 
    
    tree = nltk.ne_chunk(post_sent[0])

    result = []
    for subtree in tree.subtrees(): 
        #print(subtree)
        # Check for labels as PERSON in subtree
        if subtree.label() == 'PERSON': 
            entity = ""         
            for leaf in subtree.leaves(): 
                entity = entity + leaf[0] + " "
            result.append(entity.strip())

    # there are few names which has been tagged as GPE instead of PERSON.        
    if result == []:
        for subtree in tree.subtrees(): 
            #print(subtree)
            # Check for labels as GPE in subtree
            if subtree.label() == 'GPE': 
                entity = "" 
                for leaf in subtree.leaves(): 
                     entity = entity + leaf[0] + " "
                result.append(entity.strip()) 
        
        return result[0]
    
    # Join first and last name
    result = ' '.join(list(set(result)))
    
    # return the name of the player
    return result

def country_of_origin(doc):
    
    result=[]
    
    # regex pattern to search string "is a" or "is an"
    
    g=re.search('is a\s\w+\s',doc)
    
    # if pattern found, then 'g' will return entire string like "is a Brazilian"
    
    if(g!=None):
        grp=g.group(0)
        
        #remove stop words 
        
        result.append(grp.replace("is a", ""))
    else:
      
        # if pattern found with "is an Argentine"
        g=re.search('is an\s\w+\s',doc)
        grp=g.group(0)
        
        #remove stop words and copy the result in output file
        result.append(grp.replace("is an", ""))
    return result

def date_of_birth(doc):
    #code goes here
    
    # below pattern represents for example "born 5 February 1985"
    pattern = re.compile(r'\bborn\b\s\d+\s[A-Z]\w+\s\d+')
    for match in re.finditer(pattern, doc):
        matching_string = match.group(0)
    
    # remove "born" from output date
    date = matching_string.replace("born ","")
    return date

def team_of_the_player(doc):
    
    team =[]
    
    # below pattern represents example "for Spanish club Arsenal team and Brazilian national team"
    pattern = re.compile(r'\bfor\b([^.]*?team\b)', re.VERBOSE)
    
    # get the match of above pattern from doc 
    for match in re.finditer(pattern, doc):
        
        # remove stop words like 'for' or 'the' or 'and'
        matching_string = ''.join(match.groups())        
        pattern = re.compile(r'\b(' + r'|'.join(stopwords.words('english')) + r')\b\s*')
        matched = pattern.sub(',', matching_string)
        
        # Hence remove empty elements
        team.append(matched.replace(',,', ',').lstrip())
        
    # get the first list from list of list of team names
    team_list = team[0].split(',')
    
    # convert the list of list into a single list
    team_list = [name for name in team_list if name.strip()]
    return team_list

def position_of_the_player(doc):
    #code goes here
    
    # key: name of the player and 
    # value: postions he played
    dictvalue = {}
    
    # get the name of the player
    name = name_of_the_player(doc)
    
    # number of positions available actually
    positions_available = ['forward', 'midfielder', 'midfield', 
                           'striker', 'winger', 'defender', 'goalkeeper']
    position = []
    for word in positions_available:
        
        # if word from positions_available is present in doc then append the positions
        if word in nltk.word_tokenize(doc):
            position.append(word)
            
    # append Name and Positions in output variable dictvalue
    dictvalue['Name'] = name_of_the_player(doc)
    dictvalue['Positions'] = position
    return dictvalue
```

```python
# Use pyld to create jsonld view. This library is an implementation of the JSON-LD specification in Python.
from pyld import jsonld
import json
arg=[]

def generate_jsonld(arg, use_arg6 = True):
    
    #Hint: arg1,arg2,..... are the arguments you will be passing to the function
    
    # if cups_and_leagues_participated() function not active 
    if use_arg6 == False:
        doc = {
            "http://my-soccer-ontology.com/footballer/name_of_the_player":  "http://my-soccer-ontology.com/footballer/" +  name_of_the_player(list_of_doc[0]),
            "http://my-soccer-ontology.com/name": arg[0],
            "http://my-soccer-ontology.com/born": arg[2],
            "http://my-soccer-ontology.com/country": arg[1],
            "http://my-soccer-ontology.com/team": {"@id": "http://my-soccer-ontology.com/team", 
                                           "@type": arg[3]},
            "http://my-soccer-ontology.com/position": {"@id": "http://my-soccer-ontology.com/position", 
                                           "@type": arg[4]['Positions']}    
        }

        context = {
            " @id": "http://my-soccer-ontology.com/footballer/name_of_the_player",
            "name": "http://my-soccer-ontology.com/name",
            "born": "http://my-soccer-ontology.com/born",
            "country": "http://my-soccer-ontology.com/country",
            "team": {"@id": "http://my-soccer-ontology.com/team", 
                         "@type": "@id"},
            "position": {"@id": "http://my-soccer-ontology.com/position", 
                         "@type": "@id"}
        }
        # compact doc according to above context
        # see: http://json-ld.org/spec/latest/json-ld/#compacted-document-form
        compacted = jsonld.compact(doc, context)

        # after compaction remove @context key with value which would be no more required
        compacted.pop('@context')
        
        # renaming ' @id' to '@id'
        compacted['@id'] = compacted.pop(' @id')
        
        # renaming '@type' to 'type' from key 'position'
        compacted['position']['type'] = compacted['position'].pop('@type')

        # remove unnecessary special character like '/' from values like '/forward' from positions list
        if type(compacted['position']['type']) is str:
            compacted['position']['type'] = compacted['position']['type'].replace("/", "")
        else:
            compacted['position']['type'] = [y.replace('/', '') for y in compacted['position']['type']]

        # renaming '@type' to 'name' from key 'team'
        compacted['team']['name'] = compacted['team'].pop('@type')
        # remove unnecessary special character like '/' from values like '/Spanish Club team' from teams list
        if type(compacted['team']['name']) is str:
            compacted['team']['name'] = compacted['team']['name'].replace("/", "")
        else:
            compacted['team']['name'] = [y.replace('/', '') for y in compacted['team']['name']]
        
        # returning json format of output file
        return json.dumps(compacted, indent=2)
    
    # if cups_and_leagues_participated() function is active, do the same with extra argument of 
    # cups_and_leagues_participated() output 
    else:    
        doc = {
            "http://my-soccer-ontology.com/footballer/name_of_the_player":  "http://my-soccer-ontology.com/footballer/" +  name_of_the_player(list_of_doc[0]),
            "http://my-soccer-ontology.com/name": arg[0],
            "http://my-soccer-ontology.com/born": arg[2],
            "http://my-soccer-ontology.com/country": arg[1],
            "http://my-soccer-ontology.com/team": {"@id": "http://my-soccer-ontology.com/team", 
                                           "@type": arg[3]},
            "http://my-soccer-ontology.com/position": {"@id": "http://my-soccer-ontology.com/position", 
                                           "@type": arg[4]['Positions']},
            "http://my-soccer-ontology.com/leagues": {"@id": "http://my-soccer-ontology.com/leagues", 
                                           "@type": arg[5]}
        }

        context = {
            " @id": "http://my-soccer-ontology.com/footballer/name_of_the_player",
            "name": "http://my-soccer-ontology.com/name",
            "born": "http://my-soccer-ontology.com/born",
            "country": "http://my-soccer-ontology.com/country",
            "team": {"@id": "http://my-soccer-ontology.com/team", 
                         "@type": "@id"},
            "position": {"@id": "http://my-soccer-ontology.com/position", 
                         "@type": "@id"},
            "leagues": {"@id": "http://my-soccer-ontology.com/leagues", 
                         "@type": "@id"},
        }
        compacted = jsonld.compact(doc, context)
        #flattened = jsonld.flatten(doc)
        compacted.pop('@context')
        compacted['@id'] = compacted.pop(' @id')
        
        compacted['position']['type'] = compacted['position'].pop('@type')
        if type(compacted['position']['type']) is str:
            compacted['position']['type'] = compacted['position']['type'].replace("/", "")
        else:
            compacted['position']['type'] = [y.replace('/', '') for y in compacted['position']['type']]

        compacted['leagues']['type'] = compacted['leagues'].pop('@type')
        if type(compacted['leagues']['type']) is str:
            compacted['leagues']['type'] = compacted['leagues']['type'].replace("/", "")
        else:
            compacted['leagues']['type'] = [y.replace('/', '') for y in compacted['leagues']['type']]

        compacted['team']['name'] = compacted['team'].pop('@type')
        if type(compacted['team']['name']) is str:
            compacted['team']['name'] = compacted['team']['name'].replace("/", "")
        else:
            compacted['team']['name'] = [y.replace('/', '') for y in compacted['team']['name']]

        return json.dumps(compacted, indent=2)
        
print("JSON-LD Format is: \n", generate_jsonld([arg1,arg2,arg3,arg4,arg5], use_arg6 = False))
```

# Task 6 
Identifying one other relation (besides team and player) and write a function to extract this. Also extending the JSON-LD output accordingly.

```python
# below is the function for various cups or league particpated by each player
def cups_and_leagues_participated(doc):
    
    doc_no=[]
    team =[]
    pattern = re.compile(r'''([A-Z][a-z]*\s\bLeague\b)|                 # to get pattern like "Champions League"
                         \b[A-Z]*\b\s([A-Z][a-z]*\s\bLeague\b)|         # to get pattern like "UEFA Champions League"
                         \b[A-Z]*\b\s([A-Z][a-z]*\s\bCups\b)|           # to get pattern like "World Cups"
                         \b[A-Z]*\b\s([A-Z][a-z]*\s\bChampionships\b)|  # to get pattern like "European Championships"
                         ([A-Z][a-z]*\s\bCup\b)''', re.VERBOSE)         # to get pattern like "World Cup"
    for match in re.finditer(pattern, doc):
        matching_string = match.groups()  
        
        # remove unknown values like None
        matching_string = [name for name in matching_string if name is not None]
        team.append(matching_string)
    
    # flatten the output list
    team = [item for sublist in team for item in sublist]
    return list(set(team))

arg6 = cups_and_leagues_participated(list_of_doc[doc_no])

print("Leagues Appeared: ",arg6)
print("JSON-LD Format is: \n", generate_jsonld([arg1,arg2,arg3,arg4,arg5,arg6], use_arg6 = True))
```
